```python
#Plotting feature importance plot
def plot_feature_importance(importance,names,model_type):

#Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)

#Creating a DataFrame 
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

#Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

#Plot bar chart
    plt.figure(figsize=(10,8))

    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])

    plt.title(model_type + ' FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
```

```python
# filtering january data from df
jan_data = df[(df["YearMonth"]==201301) | (df["YearMonth"]==201401) | (df["YearMonth"]==201501) | (df["YearMonth"]==201601) | (df["YearMonth"]==201701)]
```

```python
# Function for average growth rate 
def table (col):
    table= pd.pivot_table(jan_data, values = [col], index =['Agency',"SKU"],
                         columns =['YearMonth'], aggfunc = np.mean )    
    growth_rate = np.exp(np.diff(np.log(table))) - 1
    gr_rate = pd.DataFrame(growth_rate)
    gr_rate1 = pd.DataFrame(gr_rate.sum(axis=1)/4)
    table1 = pd.pivot_table(jan_data, values = col, index =['Agency',"SKU"],
                         columns =['YearMonth'], aggfunc = np.mean ).reset_index()
    gr_rate1[2017] = table1[201701]
    gr_rate1["rate"]=gr_rate1[2017]*gr_rate1[0]
    gr_rate1['final']=gr_rate1['rate']+gr_rate1[2017]
    table1[col] = gr_rate1["final"]
    return table1
```
